```python
#@title Default title text
from Bio import SeqIO

seen = set()
records = []

for record in SeqIO.parse("/content/sequences_all.fasta", "fasta"):  
    if record.seq not in seen:
        seen.add(record.seq)
        records.append(record)


SeqIO.write(records, "OUTPUT-FILE.fasta", "fasta")
```
